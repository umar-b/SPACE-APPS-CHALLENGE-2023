```python
df_raw['DateTime'] = pd.to_datetime(df_raw['DateTime'])

# Set the DateTime column as the index
df_raw.set_index('DateTime', inplace=True)

# Resample the data into 3-hour intervals and calculate the mean
raw_mean = df_raw.resample('3H').mean()

# Print the resulting mean data
print(raw_mean)

# Create a DataFrame with custom column names
#df_raw_mean = pd.DataFrame(raw_mean, columns=custom_column_names)

#print(df_raw_mean)
```

Getting Kp values from a specific timeframe from here https://kp.gfz-potsdam.de/en/data:

```python
from getKpindex import getKpindex
(time, index, status) = getKpindex('2023-01-01', '2023-05-02', 'Kp')
```

formatting time and index and then pairing them:

```python
from datetime import datetime

date_strings = time

formatted_dates = []

for date_string in date_strings:
    # Parse the input date string into a datetime object
    date_obj = datetime.strptime(date_string, '%Y-%m-%dT%H:%M:%SZ')
    
    # Format the datetime object as a string in the desired format
    formatted_date = date_obj.strftime('%Y-%m-%d %H:%M:%S')
    
    # Append the formatted date to the list
    formatted_dates.append(formatted_date)

#print(formatted_dates)


index_list = list(index)
#print(index_list)

# Pair formatted_dates with values
date_Kp_pairs = list(zip(formatted_dates, index_list))

# Print the pairs
for pair in date_Kp_pairs:
    print(pair)
```

Second attempt at correlation raw data and Kp-values:

```python
df_Kp = pd.DataFrame(date_Kp_pairs, columns=['DateTime', 'Kp-value'])
#print(df_Kp)

df_raw_mean['DateTime'] = pd.to_datetime(raw_mean['DateTime'])
df_Kp['DateTime'] = pd.to_datetime(df_Kp['DateTime'])

#print(df_Kp)
#print(df_raw)
merged_df = pd.merge(df_raw_mean, df_Kp, on='DateTime', how='inner')
#print(merged_df)
correlation_matrix = merged_df.corr()
#print(correlation_matrix)

# Extract correlations of raw data columns and Kp values
correlations_with_Kp_values = correlation_matrix.iloc[54, 1:53]
#print(correlations_with_Kp_values)

# Create a DataFrame from the selected row
row_df = pd.DataFrame(correlations_with_Kp_values)

# Transpose the DataFrame so that it's a single row with columns as the index
row_df = row_df.transpose()

# Plot the selected row as a heatmap
plt.figure(figsize=(17, 2))  # Adjust the figure size as needed
sns.heatmap(row_df, cmap='coolwarm', annot=False, fmt=".2f")  # Adjust the colormap and other settings as needed
plt.title('Correlation Row Heatmap')
plt.show()
```
